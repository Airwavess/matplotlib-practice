```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-3, 3, 50)
y1 = x * 2 + 1
y2 = x ** 2
```

# plt.figure()

`matplotlib` 的 figure 就是一個單獨的數字小窗口，小窗口裡面還可以有更多的小圖片。

```python
plt.figure()
plt.plot(x, y1)

plt.figure(num=3, figsize=(8, 5))
plt.plot(x, y2)
plt.plot(x, y1, color='red', linewidth=2.0, linestyle='--')
plt.show()
```
